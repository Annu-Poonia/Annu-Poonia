### **Data distribution**
Here we want to try to answer the most basic questions about how the data is distributed.

#### **Top and bottom rank countries in World Happiness Report 2021**

```python
a= df1.head(5)
b =df1.tail(5)
```

```python
plt.figure(figsize=(11,3))
sns.barplot(x=a['Ladder score'], y=a['Country name']).set(title='Top 5 countries in Happiness index')
```

```python
plt.figure(figsize=(11,3))
sns.barplot(x=b['Ladder score'], y=b['Country name']).set(title='Bottom 5 countries in Happiness index')
```

We can see that ladder score is a variable to calculate happiness and it is considered as happiness index.Countries with highest ladder score are happiest and countries with lowest ladder score are most upset.<br/>

#### **Happiness score in various regions of the world(2021)**

```python
fig = px.pie(df1, values='Ladder score', names='Regional indicator', title='Happiness score in various regions of the world')
fig.show()
```

From this analysis we can say that,**Sub-Saharan Africa** region is having highest percentage of happiness.<br/>

#### **Correlation of social support on happiness**

```python
plt.figure(figsize=(17,6))
fig1 =px.scatter(df1, x = 'Ladder score', y = 'Social support', range_y = [0,1.2],
        color = 'Regional indicator', 
        labels = {'Social support': 'social support', 'Ladder score': 'happiness score', 'Regional indicator': 'name of the region'},
        hover_name = 'Country name', size="Ladder score",size_max=20)
        
fig1.update_traces(marker_size=12)
fig1.show()
```

We can clearly see that there is **positive correlation** between happiness and social support.Countries with good social support are relatively happy than that of countries with lesser social support.<br/>

#### **Correlation of freedom to make life choices vs happiness score**

```python
plt.figure(figsize=(11,5))
fig2 =px.scatter(df1, x = 'Ladder score', y = 'Freedom to make life choices', range_y = [0,1.2],
        color = 'Regional indicator', 
        labels = {'Freedom to make life choices': 'Freedom to make life choices', 'Ladder score': 'happiness score', 'Regional indicator': 'name of the region'},
        hover_name = 'Country name',hover_data = {'Regional indicator': False, 'Ladder score': False, 'Freedom to make life choices': True},
        color_discrete_sequence=["red", "blue","green","orange","purple","yellow","teal","grey","cyan","violet"])
fig2.update_traces(marker_size=12)
fig2.show()
```

We can clearly see that there is **positive correlation** between happiness and freedom to make life choices.Countries which give more freedom to make life choices are relatively happier.<br/>

#### **Features that are contributing towards happiness**

```python
df3=df1[['Country name', 'Regional indicator', 'Ladder score',
        'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']]
```

```python
x=df3.corr()['Ladder score'].sort_values(ascending=False)

plt.figure(figsize=(18,13))
plt.suptitle("Features that are contributing towards happiness",family='Serif', weight='bold', size=21)
j=0
for i in x.index[1:]:
    
    c=plt.subplot(421+j)
    c=sns.regplot(data=df1, x='Ladder score',y=i,color="y",marker="+")
    c.legend('')
    j=j+1
plt.legend('')
```

* **Logged GDP per capita**: Correlation between GDP per capita and happiness is positive.Countries with good GDP score are happiest.<br/>
* **Healthy life expectancy**: Correlation between Healthy life expectancy and happiness is positive.Countries with overall good health are relatively happier.<br/>
* **Social support**: There is positive correlation between happiness and social support.Countries with good social support are relatively happy than that of countries with lesser social support.<br/>
* **Freedom to make life choices**: We can clearly see that countries which give more freedom to make life choices are relatively happier.<br/>
* **Generosity**: There is no correlation between generosity and happiness score or we can say that there is no relation between kindness and happiness in World happiness report 2021.<br/>
* **Perceptions of corruption**: There is negative correlation between perceptions of corruption and happiness or we can say that most corrupted countries are less happy and vive versa.<br/>

#### India vs top 5 happiest countries(2021)

```python
a = df2.groupby('Country name')['Life Ladder'].mean().nlargest(5)
plt.figure(figsize=(16,5))
sns.lineplot(data=df2[df2['Country name'].isin(a.index)], x='year',y='Life Ladder',hue="Country name")
sns.lineplot(data=df2[df2['Country name']=='India'], x='year',y='Life Ladder', linewidth=3, ls='dotted')
```

#### **Historical happiness trend of India(2006-2020)**

```python
ind=df2[df2['Country name']=='India']
ind
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=ind,x="year",y="Life Ladder",marker='o',color='r')
plt.title("Life Ladder")
plt.show()
```

There are certain peaks with regards to India's ladder score as seen in the years **2008, 2010, 2012 and 2020**. There are also certain drops as seen in the years 2007, 2009, 2011 and an overall low in 2019. A continouse decrease can be seen from 2012 to 2019. **The overall trend for the ladder score is decreasing**.<br/>

#### **Features that are contributing towards happiness(India)(2006-2020)**


```python
plt.figure(figsize=(10,4))
sns.lineplot(data=ind,x="year",y="Log GDP per capita",marker='o',color='r')
plt.title("Log GDP per capita")
plt.show()

plt.figure(figsize=(10,4))
sns.lineplot(data=ind,x="year",y="Social support",marker='o',color='g')
plt.title("Social support")
plt.show()

plt.figure(figsize=(10,4))
sns.lineplot(data=ind,x="year",y="Healthy life expectancy at birth",marker='o',color='orange')
plt.title("Healthy life expectancy at birth")
plt.show()

plt.figure(figsize=(10,4))
sns.lineplot(data=ind,x="year",y="Freedom to make life choices",marker='o',color='purple')
plt.title("Freedom to make life choices")
plt.show()

plt.figure(figsize=(10,4))
sns.lineplot(data=ind,x="year",y="Generosity",marker='o',color='y')
plt.title("Generosity")
plt.show()

plt.figure(figsize=(10,4))
sns.lineplot(data=ind,x="year",y="Perceptions of corruption",marker='o')
plt.title("Perceptions of corruption")
plt.show()
```

**Features that are contributing towards happiness(analysis)**:
* India's GDP is on an **increasing trend** over the years, with the exception of 2020.<br/>
* There are some distinct **lows in social support** 2007, 2012 and 2019, 2012 being the lowest and some distinct highs in 2008, 2014 and  2018. **India's social support is on a decreasing trend over the years**.<br/>
* Healthy life expectancy at birth is on a decreasing trend over the years.<br/>
* There's a **slight increasing trend in freedom to make life choices**, but there are distincts** drops as seen in years 2008, 2012, 2015 and 2019**. With an all time **low in year 2012**.There are certain peaks with regards to India's Freedom to make life choices as seen in the years 2011, 2014, 2017 and 2020.<br/>
* **Generosity amongst Indians have increased** over time, with spikes in 2010, 2013, 2016 and 2019 and an all time low in 2008.<br/>
* There were **peaks in corruption in 2011, 2014 and 2018**, and the perception of corruption has decreased over the years.<br/>
