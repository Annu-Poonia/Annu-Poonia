<a id="process-code-concatenating"></a>
**<font size="5">Import files and merge the dataset</font>**

**All the csvs files will be concatenated into one dataframe.**

```python
datasets = """/kaggle/input/tripdata/202004-divvy-tripdata/202004-divvy-tripdata.csv
/kaggle/input/tripdata/202005-divvy-tripdata/202005-divvy-tripdata.csv
/kaggle/input/tripdata/202006-divvy-tripdata/202006-divvy-tripdata.csv
/kaggle/input/tripdata/202007-divvy-tripdata/202007-divvy-tripdata.csv
/kaggle/input/tripdata/202008-divvy-tripdata/202008-divvy-tripdata.csv
/kaggle/input/tripdata/202009-divvy-tripdata/202009-divvy-tripdata.csv
/kaggle/input/tripdata/202010-divvy-tripdata/202010-divvy-tripdata.csv
/kaggle/input/tripdata/202011-divvy-tripdata/202011-divvy-tripdata.csv
/kaggle/input/tripdata/202012-divvy-tripdata/202012-divvy-tripdata.csv
/kaggle/input/tripdata/202101-divvy-tripdata/202101-divvy-tripdata.csv
/kaggle/input/tripdata/202102-divvy-tripdata/202102-divvy-tripdata.csv
/kaggle/input/tripdata/202103-divvy-tripdata/202103-divvy-tripdata.csv""".split("\n")

dfs = [ pd.read_csv(f, engine="c") for f in datasets]
df = pd.concat(dfs, ignore_index=True)
del dfs
```

**<font size="4">Change the data type of columns(started_at and ended_at)</font>**

```python
df["ended_at"] = pd.to_datetime(df["ended_at"])
df["started_at"] = pd.to_datetime(df["started_at"])
df.info()
```

```python
df["trip_duration"] = (df["ended_at"] - df["started_at"])
df["day_of_week"] = df["started_at"].dt.day_name()
df["trip_month"] = df["started_at"].dt.month_name()
df[:5]
```

<a id="analyze-code-data"></a>
### **Data distribution**
Here we want to try to answer the most basic questions about how the data is distributed.

**<font size="4">Average ride count and average trip duration per member type</font>**
#### **Casuals vs members**
How much of the data is about members and how much is about casuals?

```python
df1 = df.loc[ (df["trip_duration"] > pd.Timedelta(0)) ] \
    .groupby( ['member_casual'], as_index=False ) \
    .agg({'trip_duration':'mean','ride_id':'count'}) \
    .sort_values('trip_duration', ascending=False) \
    .rename(columns={"trip_duration":"avg_trip_duration", "ride_id":"count"})

df1[:5]
```

 **59%** users are annual members whereas **41%** are casual riders.
 
 casual rider's average trip duration is much higher than that of annual members: **44 minutes vs 16 minutes**.

**<font size="4">Calculate the type of preferred bike per member type</font>**

```python
df2 = df.loc[ (df["trip_duration"] > pd.Timedelta(0)) ] \
    .groupby( ['member_casual', 'rideable_type'], as_index=False ) \
    .agg({'ride_id':'count','trip_duration':'mean'}) \
    .sort_values('ride_id', ascending=False) \
    .rename(columns={"rideable_type":"rideable_type_count", "trip_duration":"avg_trip_duration","ride_id":"ride_count"}) 
    
df2[:6]
```

**It's important to note that:**
* Docked bikes have the biggest volume of rides, but this can be that the company may have more docked bikes.
* Members have a bigger preference for classic bikes, 56% more.
* Casual riders use less docked bikes,electric_bike and classic_bike than annual members.

**<font size="4">Calculate the number of rides by day of the week per member type</font>**

```python
df3 = df.loc[ (df["trip_duration"] > pd.Timedelta(0))] \
      .groupby(['member_casual','day_of_week'],as_index=False) \
      .agg({'ride_id':'count','trip_duration':'mean'}) \
      .sort_values('ride_id',ascending=False) \
      .rename(columns={"ride_id":"ride_count","trip_duration":"avg_trip_duration"})
df3[:15]
```

**Saturday** is by far the day with the most rides for casual riders **(335034)**, followed by **Sunday**. It seems to be the case for annual members too, however ride number volumes are much more evenly distributed throughout all the days of the week for them.

**For both member types weekend rides have the longest duration**. However, standard deviation for casual riders is larger compared to that members.

* The biggest volume of data is on the weekend.
* Saturday has the biggest data points
* Members may have the biggest volume of data, besides on saturday. On this weekday, casual take place as having most data points.

**<font size="4">Calculate ride count and average ride duration(monthly)(casual)</font>**

```python
df4 = df.loc[ (df["trip_duration"] > pd.Timedelta(0)) \
             & (df["member_casual"] == 'casual') ] \
    .groupby( ['trip_month'], as_index=False ) \
    .agg({'ride_id':'count','trip_duration':'mean'}) \
    .sort_values('ride_id', ascending=False) \
    .rename(columns={"ride_id":"ride_count"})
df4[:12]
```

**<font size="4">Calculate ride count and average ride duration(monthly)(member)</font>**

```python
df5 = df.loc[ (df["trip_duration"] > pd.Timedelta(0)) \
             & (df["member_casual"] == 'member') ] \
    .groupby( ['trip_month'], as_index=False ) \
    .agg({'ride_id':'count','trip_duration':'mean'}) \
    .sort_values('ride_id', ascending=False) \
    .rename(columns={"ride_id":"ride_count"})
df5[:12]
```

#### **Monthly evolution of number of rides and average ride duration by member type:**

In both cases (annual members and casual riders), **we appreciate a certain seasonality in terms of numbers of rides which reaches its peak in the summer months**. Average ride duration tends to follow the same seasonality evolution. This is obviously due to weather conditions and does not have a specific impact on establishing differences between casual riders and annual members.

**<font size="4">Calculate start and end station for casual members(weekends):</font>**

```python
df5 = df.loc[ (df["member_casual"]=='casual')
             & (df["day_of_week"].isin(['Saturday','Sunday']))
             & (df["trip_duration"] > pd.Timedelta(0))
            ] \
    .groupby( ['start_station_name', 'end_station_name'], as_index=False ) \
    .agg({'ride_id':'count','trip_duration':'mean'}) \
    .sort_values('ride_id', ascending=False) \
    .rename(columns={"ride_id":"ride_count", "trip_duration":"avg_trip_duration"})
df5[:5]
```

In almost results start and end stations are same, this clearly indicates that riders use bikes for **recreational purposes**.

**<font size="4">Average duration and ride count of rides with same start and end station </font>**

```python
df6 = df.loc[ (df["start_station_name"] == df["end_station_name"])
             & (df["trip_duration"] > pd.Timedelta(0))] \
            .agg({'ride_id':'count','trip_duration':'mean'}) \
            .rename({"ride_id":"rideable_count", "trip_duration":"avg_trip_duration"})


df6[:20]
```

**<font size="4">Average duration and ride count of rides with different start and end station </font>**

```python
df7 = df.loc[ (df["start_station_name"] != df["end_station_name"])
             & (df["trip_duration"] > pd.Timedelta(0))] \
            .agg({'ride_id':'count','trip_duration':'mean'}) \
            .rename({"ride_id":"rideable_count", "trip_duration":"avg_trip_duration"})
df7[:20]
```

#### **Average duration of rides with same start and end station vs average duration of rides with different start and end stations:**

Rides with the same start and end station have a longer average duration: **50 minutes vs 25**.
